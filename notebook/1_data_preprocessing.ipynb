```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import matplotlib.dates as mdates
```

```python
# 加载数据
data = pd.read_csv('../data/raw/stock_dataset_2.csv')
```

```python
# 确保日期列为datetime类型
data['date'] = pd.to_datetime(data['date'])
```

```python
# 数据集划分
train_data, test_data = train_test_split(data, test_size=0.2, shuffle=False)
train_data, val_data = train_test_split(train_data, test_size=0.25, shuffle=False)
```

```python
# 保存处理后的数据
train_data.to_csv('../data/processed/train_data.csv', index=False)
val_data.to_csv('../data/processed/val_data.csv', index=False)
test_data.to_csv('../data/processed/test_data.csv', index=False)
```

```python
# 数据可视化
plt.figure(figsize=(14,7))
plt.plot(data['date'], data['close'], label='Close Price')

# 设置日期格式和间隔
plt.gca().xaxis.set_major_locator(mdates.YearLocator(2))  # 每两年一个主刻度
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
